```python
# Memanggil module os untuk membuat dan menambahkan file
import os

# Memanggil module collections untuk mempermudah pembuatan dictionary
import collections
```

```python
# Assign path ke directory komputer

download_path_directory = r"D:\Download_sim"
```

```python
# Mapping

## Membuat dictionary file dan tipe file
file_map = collections.defaultdict()

for file_name in os.listdir(download_path_directory):

    ## Memisahkan nama file dan tipe file, dan mengambil hanya tipenya saja
    file_type = file_name.split('.')[-1]

    ## Memmasukkan tipe file ke dictionary sebagai key, dan memasukkan file ke dalam tipe file
    file_map.setdefault(file_type, []).append(file_name)
    
## Cek pembuatan dictionary
print(file_map)
```

```python
## Alternatif cek pembuatan dictionary: metode lebih rapih

### Import pprint
from pprint import pprint

## Print alternatif
pprint(file_map)
```

```python
# Pembuatan folder pada directory download path
for folder_name_type, folder_files in file_map.items():
    folder_path = os.path.join(download_path_directory, folder_name_type)

    # Membuat folder jika tidak ada pada path
    if not os.path.exists(folder_path):
        os.mkdir(folder_path)

    # Memasukkan file ke masing-masing folder tipe berdasarkan tipe
    for folder_files in folder_files:

        # Define sumber dan tujuan pemasukkan file
        source = os.path.join(download_path_directory, folder_files)
        destination = os.path.join(folder_path, folder_files)
    
        # Simulasi pemindahan file menggunakan print
        pprint(f'Moving {source} to {destination}')

        # Memindahkan file ke folder
        os.rename(source, destination)
        
# Print
print("\n")
print('Your file has been sorted! :)')
```
